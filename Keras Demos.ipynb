```python
import pandas as pd
```

```python
data = pd.read_csv("/content/sample_data/california_housing_train.csv")
```

```python
test_data = pd.read_csv("/content/sample_data/california_housing_test.csv")
```

```python
data
```

```python
data = (data - data.mean())/data.std()
```

```python
data
```

```python
from sklearn.model_selection import train_test_split
```

```python
Y = data['median_house_value']

X = data.iloc[:,0:8]
```

```python
import numpy as np
```

```python
X = np.array(X)
Y = np.array(Y)
```

```python
X_train, X_cv, Y_train, Y_cv = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
m = X_train.shape[1]
```

```python
X_train.shape
```

```python
Y_train.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense,InputLayer,BatchNormalization
from keras.regularizers import l2
```

```python
def create_model():

  network = Sequential()
  network.add(InputLayer(input_shape=(m,)))
  network.add(Dense(units=X_train.shape[1],activation="relu",kernel_initializer="glorot_normal",kernel_regularizer="l2"))
  network.add(BatchNormalization())
  network.add(Dense(units=X_train.shape[1],activation="relu",kernel_initializer="glorot_normal",kernel_regularizer="l2"))
  network.add(BatchNormalization())
  network.add(Dense(units=1))
  return network
```

```python
network = create_model()
```

```python
network.summary()
```

```python
network.compile(optimizer="adam",loss="mse")
```

```python
network.fit(x=X_train,y=Y_train,batch_size=136,epochs=50,validation_data=(X_cv,Y_cv))
```
