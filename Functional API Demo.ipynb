```python
from keras.models import Model
from keras.layers import Input,Flatten,Dense,BatchNormalization
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("/content/sample_data/mnist_train_small.csv")
```

```python
data = np.array(data)
```

```python
class_labels = data[:,0]

data = data[:,1:]
```

```python
class_labels = np.eye(46,46)[class_labels]
```

```python
def build_network():

  inputs_to_nw = Input(shape=(784,))

  hidden_layer1_output = Dense(units=784,activation="relu") (inputs_to_nw)

  hidden_layer2_output = Dense(units=784,activation="relu") (hidden_layer1_output)

  modified_hidden_layer2_output = BatchNormalization() (hidden_layer2_output)

  output_layer_output = Dense(units=46,activation="softmax") (modified_hidden_layer2_output)

  network = Model(inputs=inputs_to_nw,outputs=output_layer_output)

  network.compile(loss="categorical_crossentropy",metrics=["accuracy"])

  return network
```

```python
network = build_network()
```

```python
network.fit(x=data,y=class_labels,batch_size=32)
```
