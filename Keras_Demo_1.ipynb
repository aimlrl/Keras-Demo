```python
import numpy as np
```

```python
Y_train = np.random.randint(0,10,10)
```

```python
Y_train
```

```python
X_train = np.array(X_train)
X_cv = np.array(X_cv)
Y_train = np.array(Y_train)
Y_cv = np.array(Y_cv)
```

```python
from keras.layers import InputLayer,Dense,BatchNormalization
from keras.models import Sequential
```

```python
from keras.regularizers import l2
from keras.optimizers import Adam
from keras.activations import PReLU
from keras import initializers
```

```python
def build_network():

    network = Sequential()
    network.add(InputLayer(input_shape=(X_train.shape[1],)))
    network.add(Dense(units=X_train.shape[1],activation=PReLU(),kernel_initializer="glorot_normal",kernel_regularizer=l2(0.001)))
    network.add(BatchNormalization())
    network.add(Dense(units=X_train.shape[1],activation=PReLU(),kernel_initializer="glorot_normal",kernel_regularizer=l2(0.001)))
    network.add(BatchNormalization())
    network.add(Dense(units=X_train.shape[1],activation=PReLU(),kernel_initializer="glorot_normal"))
    network.add(BatchNormalization())
    network.add(Dense(units=10,activation="softmax",kernel_initializer="glorot_normal"))

    return network
```

```python
network = build_network()
```

```python
network.compile(optimizer=Adam(learning_rate=0.01),loss="mse")
```

```python
performance_history = network.fit(x=X_train,y=Y_train,batch_size=136,epochs=500,validation_data=(X_cv,Y_cv))
```

```python
training_loss = performance_history.history['loss']
cv_loss = performance_history.history['val_loss']
```

```python
epochs = range(1,len(training_loss)+1)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(epochs,training_loss,'g',label="Training Loss")
plt.plot(epochs,cv_loss,'b',label="Cross Validation Loss")
plt.title("Training and Cross Validation Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.show()
```
